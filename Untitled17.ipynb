```python
GPS.Rating=GPS.Rating.fillna(GPS['Rating'].median())#fill null values with median of Rating
```

```python
GPS.dropna(inplace=True)
```

```python
GPS.isna().sum()
```

```python
GPS.Category.unique()
```

```python
GPS.drop(GPS[GPS['Category']== '1.9'].index)
```

```python
GPS[GPS['Size']== 'Varies with divices']=np.nan
```

```python
GPS['Size']=GPS.Size.str.replace('M','e3')
#convert K value
GPS['Size']=GPS.Size.str.replace('K','e0')
```

```python
GPS['Size']=pd.to_numeric(GPS['Size'],errors='coerce')
#replace Nan value with mean
GPS['Size']=GPS['Size'].fillna(GPS['Size'].mean())
```

```python
GPS['Installs']=GPS.Installs.str.replace('+','')
GPS['Installs']=GPS.Installs.str.replace(',','')
GPS['Installs']=GPS['Installs'].astype(int)#convert to numeric
```

```python
GPS['Price']=GPS.Price.str.replace('$','')
```

```python
GPS['Price']=GPS['Price'].astype(float)
```

```python
GPS['Reviews']=GPS['Reviews'].astype(int)
#remove reviews greater than installs
GPS.drop(GPS[GPS['Reviews']>GPS['Installs']].index,axis=0,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

GPS['App']=le.fit_transform(GPS['App'])
```

```python
GPS=pd.concat([GPS,pd.get_dummies(GPS['Category'],prefix='cat')],axis=1)
```

```python
GPS.drop(['Category'],axis=1,inplace=True)
```

```python
GPS['Genres']=le.fit_transform(GPS['Genres'])
```

```python
GPS['Content Rating']=le.fit_transform(GPS['Content Rating'])
```

```python
GPS['Type']=pd.get_dummies(GPS['Type'])
```

```python
GPS['Android Ver']=le.fit_transform(GPS['Android Ver'])
```

```python
GPS['Current Ver']=le.fit_transform(GPS['Current Ver'])
```

```python
GPS['Last Updated']=le.fit_transform(GPS['Last Updated'])
```

```python
GPS.head()
```

```python
GPS.Reviews=np.log1p(GPS['Reviews'])
GPS.Size=np.log1p(GPS['Size'])
GPS.Installs=np.log1p(GPS['Installs'])
GPS.App=np.log1p(GPS['App'])
GPS['Android Ver']=np.log1p(GPS['Android Ver'])
GPS['Last Updated']=np.log1p(GPS['Last Updated'])
GPS['Current Ver']=np.log1p(GPS['Current Ver'])
```

```python
GPS=np.log1p(GPS)
```

```python
GPS.head()
```

```python
X=GPS.drop(['Rating'],axis=1)
Y=GPS.Rating
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=10)
```

```python
from sklearn.neighbors import KNeighborsRegressor
lm=KNeighborsRegressor(n_neighbors=40)
lm.fit(X_train,Y_train)
```

```python
Y_pred=lm.predict(X_test)
```

```python
from sklearn import metrics
print('MAE:',metrics.mean_absolute_error(Y_test,Y_pred))
print('MSE:',metrics.mean_squared_error(Y_test,Y_pred))
print('RMSE:',np.sqrt(metrics.mean_absolute_error(Y_test,Y_pred)))
```

```python
from sklearn.ensemble import RandomForestRegressor
rn=RandomForestRegressor(n_estimators=500)
rn.fit(X_train,Y_train)
```

```python
rn_pred=rn.predict(X_test)
```

```python
print('MAE:',metrics.mean_absolute_error(Y_test,rn_pred))
print('MSE:',metrics.mean_squared_error(Y_test,rn_pred))
print('RMSE:',np.sqrt(metrics.mean_absolute_error(Y_test,rn_pred)))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gn=GradientBoostingRegressor()
gn.fit(X_train,Y_train)
```

```python
gn_pred=gn.predict(X_test)
```

```python
print('MAE:',metrics.mean_absolute_error(Y_test,gn_pred))
print('MSE:',metrics.mean_squared_error(Y_test,gn_pred))
print('RMSE:',np.sqrt(metrics.mean_absolute_error(Y_test,gn_pred)))
```
